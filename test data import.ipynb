```python
# 1. Télécharger les données historiques des ETF QQQ et VOO
tickers = ['QQQ', 'VOO', 'META', 'AMZN', 'GOOGL', 'AAPL', 'NFLX', 'UBER', 'SPOT', 'MSFT', 'NVDA']
start_date = '2018-01-01'
end_date = '2023-01-01'
data = yf.download(tickers, start=start_date, end=end_date)['Adj Close']

# 2. Calculer les rendements quotidiens
returns = data.pct_change().dropna()

# 3. Calculer les rendements cumulés
cumulative_returns = (1 + returns).cumprod()

# 4. Visualisation des rendements cumulés
plt.figure(figsize=(10, 6))
for ticker in tickers:
    plt.plot(cumulative_returns.index, cumulative_returns[ticker], label=ticker)

plt.title('Rendements cumulés de QQQ et VOO')
plt.xlabel('Date')
plt.ylabel('Valeur cumulée')
plt.legend()
plt.grid(True)
plt.show()

# 5. Calcul des métriques de performance
for ticker in tickers:
    total_return = cumulative_returns[ticker].iloc[-1] - 1
    annualized_return = returns[ticker].mean() * 252
    annualized_volatility = returns[ticker].std() * (252 ** 0.5)
    sharpe_ratio = annualized_return / annualized_volatility
    
    print(f"--- {ticker} ---")
    print(f"Rendement total : {total_return:.2%}")
    print(f"Rendement annualisé : {annualized_return:.2%}")
    print(f"Volatilité annualisée : {annualized_volatility:.2%}")
    print(f"Ratio de Sharpe : {sharpe_ratio:.2f}")
```
